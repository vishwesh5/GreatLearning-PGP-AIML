## <font style = "color:rgb(8,133,37)">Step 4</font>

### <font style = "color:rgb(8,133,37)">Do charges of people who smoke differ significantly from the people who don't?</font>

```python
# Let's make two groups of people
# First, those who smoke and second, those who don't
insurance.groupby("smoker").mean()
```

As we can see from the above output, on average, the people who smoke have much higher charges than those who don't.

Let us now verify the same using t-test. In all cases we will use significance level 0.05.

H0: People who smoke and those who don't have the same mean charges.

Ha: People who smoke and those who don't have different mean charges.

We will assume that both groups have the same variance.

```python
from scipy.stats import ttest_ind
```

```python
group1 = insurance[insurance["smoker"]=="yes"].charges
group2 = insurance[insurance["smoker"]=="no"].charges
```

```python
t_statistic, p_value = ttest_ind(group1, group2)
print(t_statistic, p_value)
```

As the p-value < 0.05, thus we can reject the null hypothesis.

Thus, charges of people who smoke versus those who don't differ significantly.

### <font style = "color:rgb(8,133,37)">Does bmi of males differ significantly from that of females?</font>

```python
# Again, we will group the data based on sex
insurance.groupby('sex').mean()
```

There is no significant difference in the bmi of males and females.

We will again use t-test in this case. The null and alternate hypotheses are as follows:

H0: Mean bmi of males is the same as the mean bmi of females

Ha: Mean bmi of males is not the same as the mean bmi of females

```python
group1 = insurance[insurance["sex"]=="female"].bmi
group2 = insurance[insurance["sex"]=="male"].bmi
```

```python
t_statistic, p_value = ttest_ind(group1, group2)
print(t_statistic, p_value)
```

The p-value is more than 0.05 so we cannot reject the null hypothesis. Thus, we can't say that there is a significant difference in the mean bmi of males and females.

### <font style = "color:rgb(8,133,37)">Is the proportion of smokers significantly different in different genders?</font>

```python
# Select only the smokers
smokers = insurance[insurance["smoker"]=="yes"]
```

```python
# Let's see the gender distribution in smokers
smokers.sex.value_counts()
```

```python
# Proportion of males that smoke
males_smoker = smokers.sex.value_counts()["male"] * 100 / insurance.sex.value_counts()["male"]
# Proportion of females that smoke
females_smoker = smokers.sex.value_counts()["female"] * 100 / insurance.sex.value_counts()["female"]
```

```python
males_smoker,females_smoker
```

As we can see, 23.5% males are smokers in the dataset whereas only 17.37% females are smokers. There is a significant difference in the proportion.

Now let's carry out statistical analysis for this problem.

We will use proportions test for this. 

H0: Proportions of smokers are same in both genders

H1: Proportions of smokers are different in both genders

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
female_smokers = insurance[insurance["sex"]=="female"].smoker.value_counts()[1]
male_smokers = insurance[insurance["sex"]=="male"].smoker.value_counts()[1]
```

```python
insurance.sex.value_counts()
```

```python
n_females = insurance.sex.value_counts()[1]
n_males = insurance.sex.value_counts()[0]
```

```python
stat, pval = proportions_ztest([female_smokers, male_smokers] , [n_females, n_males])

if pval < 0.05:
    print(f'With a p-value of {round(pval,4)} the difference is significant. aka |We reject the null|')
else:
    print(f'With a p-value of {round(pval,4)} the difference is not significant. aka |We fail to reject the null|')
```

Thus, we can see that there is a significant difference in both proportions.

### <font style = "color:rgb(8,133,37)">Is the distribution of bmi across women with no children, one children and two children the same?</font>

```python
# Let's select all women
women = insurance[insurance["sex"]=="female"]
```

```python
# Next, let's select women with no children
women_0 = women[women["children"]==0]
# Women with 1 children
women_1 = women[women["children"]==1]
# Women with 2 children
women_2 = women[women["children"]==2]
```

```python
# Let's now plot the bmi distribution
plt.figure(figsize=(8,8))
sns.distplot(women_0.bmi,hist=False,color="red",label="No children")
sns.distplot(women_1.bmi,hist=False,color="green",label="1 children")
sns.distplot(women_2.bmi,hist=False,color="blue",label="2 children")
```

As we can see, there is not much difference in bmi distribution for women with No children and 1 children. For women with 2 children, there is a slight difference in distribution but nothing that should be considered significant.

```python
# Let's also check on the basis of 5 point summary for bmi for the 3 dataframes
women_0.bmi.describe()
```

```python
women_1.bmi.describe()
```

```python
women_2.bmi.describe()
```

```python
# Let's create a new dataframe
df1 = pd.DataFrame({'Num_Children':0,'bmi':women_0.bmi.values})
```

```python
df2 = pd.DataFrame({'Num_Children':1,'bmi':women_1.bmi.values})
df3 = pd.DataFrame({'Num_Children':2,'bmi':women_2.bmi.values})
```

```python
bmi_df = pd.DataFrame()
bmi_df = bmi_df.append(df1)
bmi_df = bmi_df.append(df2)
bmi_df = bmi_df.append(df3)
```

```python
bmi_df.head()
```

```python
bmi_df.tail()
```

```python
sns.boxplot(x = "Num_Children", y = "bmi", data = bmi_df)
plt.show()
```

A few points to note are:

1. There are much more women with No children as compared to women with 1 or 2 children. This matches our initial analysis.
2. Mean, minimum and maximum bmi for all the 3 dataframes is very close. Similarly for other quartiles.

So, we can conclude that there is no significant difference in bmi distribution across women with no children, 1 children and the two children.

Now for statistical analysis, we will use ANOVA. 

H0: Mean bmi for no children, 1 children and 2 children is same.

H1: At least one of the 3 means is different.

```python
import statsmodels.api         as     sm
from   statsmodels.formula.api import ols
```

```python
mod = ols('bmi ~ Num_Children', data = bmi_df).fit()
aov_table = sm.stats.anova_lm(mod, typ=2)
print(aov_table)
```

As the p-value is 0.79376 which is more than 0.05, we can not reject the null hypothesis. Thus, we can say that the 3 mean bmi (0 children, 1 children and 2 children) are same.
